```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
names = ["max", "min", "avg"]

# Data Load
roulette = pd.read_csv("roulette history.txt", sep = " ", names = names)
# roulette = pd.read_csv("tournament history.txt", sep = " ", names = names)
tournament = pd.read_csv("tournament history.txt", sep = " ", names = names)


# Draw Graph
plt.title("Average Fitness Courve")
plt.plot(roulette.index, roulette["avg"], label = "roulette")
plt.plot(tournament.index, tournament["avg"], label = "tournament")
plt.legend()

plt.savefig('Average Fitness Courve.png')
# mlt.
```
